```python
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

url_train = "https://raw.githubusercontent.com/drmuskangarg/Irf/refs/heads/main/train_data.csv"
train = pd.read_csv(url_train)
train.to_csv('train_irf.csv', index = False)

url_test = "https://raw.githubusercontent.com/drmuskangarg/Irf/refs/heads/main/test_data.csv"
test = pd.read_csv(url_test)
test.to_csv('test_irf.csv', index = False)

y_train_bel = train.belong
y_train_bur = train.burden
y_test_bel = test.belong
y_test_bur = test.burden
```

```python
train.head()
```

```python
from TONY.Lexicon import MarkersExtractionColab
```

```python
app = MarkersExtractionColab()
```

```python
markers_train = pd.read_csv('markers_train.csv')
markers_test = pd.read_csv('markers_test.csv')
```

```python
markers_test
```

```python
base_lr = LogisticRegression(random_state=0, max_iter=5000)
pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', base_lr),
])

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
pipe_lr.fit(markers_train, y_train_bel)
preds_lr = pipe_lr.predict(markers_test)
print('Logistic Regression')
print('Thwarted Belongingness')
print(accuracy_score(y_test_bel, preds_lr))
print(f1_score(y_test_bel, preds_lr))
print(precision_score(y_test_bel, preds_lr))
print(recall_score(y_test_bel, preds_lr))

pipe_lr.fit(markers_train, y_train_bur)
preds_lr = pipe_lr.predict(markers_test)
print(' Perceived Burdensomeness')
print(accuracy_score(y_test_bur, preds_lr))
print(f1_score(y_test_bur, preds_lr))
print(precision_score(y_test_bur, preds_lr))
print(recall_score(y_test_bur, preds_lr))
```

```python
from empath import Empath

lexicon = Empath()
empaths_features_train = [list(lexicon.analyze(text, normalize=True).values()) for text in train.text.values.tolist()]
empaths_features_test = [list(lexicon.analyze(text, normalize=True).values()) for text in test.text.values.tolist()]
```

```python
base_lr = LogisticRegression(random_state=0, max_iter=5000)
pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', base_lr),
])

print('Logistic Regression with Empath features')
pipe_lr.fit(empaths_features_train, y_train_bel)
preds_lr = pipe_lr.predict(empaths_features_test)
print('Thwarted Belongingness')
print(accuracy_score(y_test_bel, preds_lr))
print(f1_score(y_test_bel, preds_lr))
print(precision_score(y_test_bel, preds_lr))
print(recall_score(y_test_bel, preds_lr))

pipe_lr.fit(empaths_features_train, y_train_bur)
preds_lr = pipe_lr.predict(empaths_features_test)
print(' Perceived Burdensomeness')
print(accuracy_score(y_test_bur, preds_lr))
print(f1_score(y_test_bur, preds_lr))
print(precision_score(y_test_bur, preds_lr))
print(recall_score(y_test_bur, preds_lr))
```
